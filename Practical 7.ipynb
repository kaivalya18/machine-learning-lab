```python
#Explore and implement naive bayes classifier
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB
```

```python
data ={'Cheap':['Yes','Yes','No','Yes','No'],
       'Promo':['Yes','No','Yes','Yes','No'],
        'Buy?': ['Yes','Yes','No','Yes','No']
       }
```

```python
df=pd.DataFrame(data)
df
```

```python
#step 2 Encode categorical variables

le_cheap = LabelEncoder()
le_promo = LabelEncoder()
le_buy = LabelEncoder()
```

```python
df['Cheap'] = le_cheap.fit_transform(df['Cheap'])
df['Promo'] = le_promo.fit_transform(df['Promo'])
df['Buy?'] = le_buy.fit_transform(df['Buy?'])
```

```python
#step 3: Features (X) and Target (y)
X=df[['Cheap','Promo']]
y=df['Buy?']
```

```python
#step 4: train naive bayes
model=CategoricalNB()
model.fit(X,y)
```

```python
#step 5 predict on the same data
predictions=model.predict(X)
predictions
```

```python
#preform next step
df['Predictions']=predictions
df
```

```python
#show the results with decoded labels
df['Predicted_Buy?']=le_buy.inverse_transform(predictions)
df['AcutalBuy']=le_buy.inverse_transform(y)
df
```
