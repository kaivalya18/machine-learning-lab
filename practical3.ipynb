```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head()
```

```python
for i in range(4):
  df.loc[i, 'sepal length (cm)'] = np.nan
```

```python
df.head(10)
```

```python
for i in range(4,7):
  df.loc[i, 'sepal width (cm)'] = np.nan
```

```python
df.head(10)
```

```python
for i in range(7,11):
  df.loc[i, 'petal length (cm)'] = np.nan
```

```python
for i in range(7,11):
  df.loc[i, 'petal width (cm)'] = np.nan
```

```python
for col in df.columns:
  mean_value = df[col].mean()
  df[col].fillna(mean_value, inplace=True)



display(df.isnull().sum())
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

df.boxplot()
plt.title("Box Plot of Iris Dataset (Missing Values Treated)")
plt.ylabel("Measurement (cm)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

array1=[100,120,130,150,180,200,220,250]

plt.boxplot(array1)
plt.title("Box Plot of array1")
plt.ylabel("Values")
plt.show()
```
